```python
# classify 
from sklearn.cluster import KMeans
X = arr
reduced_data = np.reshape(arr[:,2],(-1,1))
landTypeCount = len(np.unique_counts(arr[:,2]).values)
print(landTypeCount)
labels = KMeans(n_clusters=len(np.unique_counts(arr[:,2]).values)).fit_predict(reduced_data)
```

```python

plt.figure(figsize=(8, 6))
scatter = plt.scatter(
    X[:, 0],
    X[:, 1],
    c=labels,
)

plt.title("2D Visualization of KMeans Clusters (PCA Projection)")
plt.show()
```
